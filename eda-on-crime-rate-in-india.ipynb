Finding out total number of victims for each year using sql query

```python
from pandasql import sqldf
sqldf('SELECT * from df where Year==2008') 
sqldf('select Year, SUM(Victims_Total) as total_victims from df group by Year')  #Total number of victims for each year
```

```python
sqldf('select SUM(Victims_Total),Year from df where Year==2008')  #Number of victims for year 2008
```

```python
sqldf('select SUM(Victims_Total) as victims, Year, Group_Name from df Group by Group_Name')
```

```python
sqldf('select Area_Name, Victims_Total, Group_Name from df where Year ==2001')
```

```python
sqldf('select Area_Name,Year, SUM(Victims_Total), Group_Name  from df group by  Year, Group_Name')
```

```python
sqldf('Select * from df Order by Victims_Total DESC')      #Data based on total victims in descending order
```

```python
sqldf("Select Area_Name, Sub_Group_Name, Year, Victims_Above_50_Yrs from df Where Victims_Above_50_Yrs >= 200")
```
