```python
df=pd.read_csv('../input/crime-in-india/32_Murder_victim_age_sex.csv')  #reading the Murder crime file 
df.head() #printing the top 5 rows using head function to check if the file has been read 
```

```python
df.fillna(0, inplace=True)
df.head(20)
```

```python
import matplotlib.pyplot as plt
plt.bar(df['Year'],df['Victims_Total'])  # Trying to observe the crime rate in India over the years
plt.title('Bar chart of Crime rate in India between 2001 & 2010')
plt.xlabel('Year')
plt.ylabel('Number of Victims')
plt.show()
```

From the above graph we can say that the highest crime rate in India is in 2001 & Lowest in 2010.


```python
df = df.sort_values('Victims_Total') #Sorting the dataset in the ascending order of total number of crimes
df.head()
```

Finding out total number of victims for each year using sql query

```python
from pandasql import sqldf
#sqldf('SELECT * from df where Year==2008')  #printign data for the year 2008
sqldf('select Year, SUM(Victims_Total) as total_victims from df group by Year')  #Total number of victims for each year
```

```python
sqldf('select SUM(Victims_Total),Year from df where Year==2008')  #Number of victims for year 2008
```

```python
sqldf('select SUM(Victims_Total) as victims, Year, Group_Name from df Group by Group_Name')
```

```python
sqldf('select Area_Name, Victims_Total, Group_Name from df where Year ==2001')
```

```python
sqldf('select Area_Name,Year, SUM(Victims_Total), Group_Name  from df group by  Year, Group_Name')
```

```python
sqldf('Select * from df Order by Victims_Total DESC')      #Data based on total victims in descending order
```

```python
sqldf("Select Area_Name, Sub_Group_Name, Year, Victims_Above_50_Yrs from df Where Victims_Above_50_Yrs >= 200")
```

```python
sqldf("Select Area_Name, Sub_Group_Name, Year, Victims_Upto_10_Yrs from df Where Victims_Upto_10_Yrs Between 20 And 40")
```
